```r
# Load package
library(scater)
library(splatter)
library(singleCellHaystack)
```

```r
set.seed(1234)
```

## check different cell numbers

```r
n_cell_time = list()
for (nCell in c(50,100,500,1000,2500,5000)){
    sce <- readRDS(file=paste('./write/SplatSim_G2_B1_sce_based_real_data_nCell_',
                       as.character(nCell),
                       '_20210903.rds',
                       sep=''));
    sce <- logNormCounts(sce);
    sce <- runUMAP(sce,ncomponents = 2,n_threads=5);
    t1 <- proc.time();
    res <- haystack(sce, assay = "counts",coord = "UMAP", method = "2D");
    t2 <- proc.time();
    t = t2-t1;
    n_cell_time[paste('nc',as.character(nCell),sep='')] <- t[3][[1]];
    write.csv(res$results,file = paste('./write/run_singleCellHaystack_SplatSim_G2_B1_sce_based_real_data_nCell_',
                       as.character(nCell),
                       '_20210903.csv',
                       sep=''))
}
#write.csv(as.data.frame(n_cell_time),file='./write/run_singleCellHaystack_SplatSim_G2_sce_based_real_data_deCells_run_time_20210909.csv')
```
