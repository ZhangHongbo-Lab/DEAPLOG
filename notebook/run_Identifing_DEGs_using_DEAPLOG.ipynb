# check different cell numbers

```python
import time
n_cell_time = dict()
for ncell in [50,100,500,1000,2500,5000]:    
    sim_count_data = pd.read_csv('./write/SplatSim_G2_B1_sce_based_real_data_counts_nCell_'+str(ncell)+'_20210903.csv',
                                 header=0,sep=',',index_col=0);
    sim_pheno_data = pd.read_csv('./write/SplatSim_G2_B1_sce_based_real_data_pheno_nCell_'+str(ncell)+'_20210903.csv',
                                 header=0,sep=',',index_col=0);
    sim_feature_data = pd.read_csv('./write/SplatSim_G2_B1_sce_based_real_data_feature_nCell_'+str(ncell)+'_20210903.csv',
                                   header=0,sep=',',index_col=0);
    sim_data = anndata.AnnData(X=sim_count_data).transpose();
    sim_data.obs = sim_pheno_data;
    sim_data.var = sim_feature_data;
    sc.pp.normalize_per_cell(sim_data, counts_per_cell_after=1e4);
    sc.pp.log1p(sim_data);
    start_t = time.time();
    markers_s = plogs.get_DEG_uniq(rdata=sim_data,adata=sim_data,group_key='Group',ratio=0,
                                     p_threshold=1,
                                     q_threshold=1,
                                  power=11)
    end_t = time.time();
    run_t = end_t-start_t;
    n_cell_time[ncell] = run_t;
    #sim_data.uns['run_PLOGS_time'] = run_t;
    #sim_data.uns['markers_s_deaplog_power11'] = markers_s;
    #sim_data.write_h5ad('./write/SplatSim_G2_sce_based_real_data_counts_nCell_'+str(ncell)+'_norm_log1p_20210903.h5ad');
    markers_s.to_csv('./write/run_plogs_power11_SplatSim_G2_B1_sce_ngene10k_ncell'+str(ncell)+'_20220107.csv');
#pd.DataFrame(n_cell_time.values(),index=n_cell_time.keys(),columns=['run_seconds']).to_csv('./write/run_plogs_power8_SplatSim_G2_B1_sce_ngene10k_De_ncells_run_time_20220107.csv')
```
